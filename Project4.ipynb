```python
x = np.linspace(-5, 5, 500)
y = norm.pdf(x, loc=0, scale=1)

z_lower = 2.56964789
z_upper = 5
x_shade = np.linspace(z_lower, z_upper, 1000)
y_shade = norm.pdf(x_shade, loc=0, scale=1)
plt.fill_between(x, y, color='lightyellow', alpha=0.5)
plt.fill_between(x_shade, y_shade, color='purple', alpha=0.5)

plt.plot(x, y)

print('Light yellow represents the rejection and purple represents the sample proportion.')
```

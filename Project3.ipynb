```python
import numpy as np

import matplotlib.pyplot as plt
%matplotlib inline

# This lets us integrate functions.
from scipy import integrate

# Special statistics things.
import scipy.stats as stats
from scipy.stats import norm

# This will let us use the "ones" and "shape" commands later.
from pylab import ones, shape
```

## 1. Finding whether or not functions below are Probability Density Functions. ##

### Part 1A: ### 
$$f(t)=\begin{cases} 10e^{-10t} & t>0, \\ 0 & t \leq 0. \end{cases}$$ 

```python
# Let's find out the area under the curve

def f1(t):
    return np.exp(-10*t) * (10)

print(integrate.quad(f1, 0, np.inf)) # Prints the result of integration, followed by the error.
print("This function satisfies the requirement for the probability function for a continous random variable.") # Rounded to 1.0
```

### Part 1B: ### 
$$f(t)=\begin{cases} 10e^{-0.1t} & t>0, \\ 0 & t \leq 0. \end{cases}$$ 

```python
def f2(t):
    return np.exp(-0.1*t) * (10)

print(integrate.quad(f2, 0, np.inf)) # Prints the result of integration, followed by the error.
print("This function does not satisfy the requirement for the probabity density function for a continous random variable.")
```

### Part 1C: ### 
$$f(t)=\begin{cases} 0.3e^{-0.3t} & t>0, \\ 0 & t \leq 0. \end{cases}$$ 

```python
def f3(t):
    return np.exp(-0.3*t) * (0.3)

print(integrate.quad(f3, 0, np.inf)) # Prints the result of integration, followed by the error.
print("This function does satisfy the requirement for the probabity density function for a continous random variable.")
```

### Part 1D: ### 
$$f(x)=\begin{cases} \frac{1}{5} & 0\leq x < 5, \\ 0 & otherwise. \end{cases}$$ 

```python
def f4(x):
    return (1/5)

print(integrate.quad(f4, 0, 5)) # Prints the result of integration, followed by the error.
print("This function does satisfy the requirement for the probabity density function for a continous random variable.") # Rounded to 1.0
```

### Part 1E: ### 
$$f(x)=\begin{cases} x & 0\leq x < \sqrt 2, \\ 0 & otherwise. \end{cases}$$ 

```python
def f5(x):
    return x

print(integrate.quad(f5, 0, np.sqrt(2))) # Prints the result of integration, followed by the error.
print("This function does satisfy the requirement for the probabity density function for a continous random variable.") # Rounded to 1.0
```

## 3. Graphing, shading, and finding area under the curve for parts 1c, 1d, and 1e. ##

#### Part C ####

```python
x1 = np.linspace(0,4,100) # Setting up starting point, ending point, and evenly spaced points
ft = np.array([f3(ti) for ti in x1]) # Getting ouput for an array of x inputs

# Diffrentiating shaded area from all area
plt.fill_between(x1, ft, color='gray', alpha=0.5, hatch = '|')
plt.fill_between(x1, ft, where=(x1 > 0.5) & (x1 < 1.25), color='red', alpha=0.5, hatch = '|')

plt.plot(x1, ft)

#Printing the area between a specified interval rounded to the nearest thousandths 
print("Below is the area under the curve from 0.5 to 1.25:")
print("{:.3f}".format(integrate.quad(f3, 0.5, 1.25)[0]))
```

## 4. Graphing, shading, and finding the area for the standard normal curve (µ = 0, σ = 1) ##

```python
x = np.linspace(-3, 3, 500)
y = norm.pdf(x, loc=0, scale=1)
plt.plot(x, y)

z_lower = -1.5
z_upper = 2.25
x_shade = np.linspace(z_lower, z_upper, 1000)
y_shade = norm.pdf(x_shade, loc=0, scale=1)
plt.fill_between(x, y, color='lightblue', alpha=0.5)
plt.fill_between(x_shade, y_shade, color='purple', alpha=0.5)

area_under_curve = stats.norm.cdf(z_upper, 0, 1) - stats.norm.cdf(z_lower, 0, 1)

print("Here is the standard normal curve with mean of 0, and standard deviation of 1.\nThe shaded region is between z = -1.5 and z = 2.25.")
print("The area under the purple shade is {:.3f}".format(area_under_curve))
```

## 5. Solving a real-life application ##

#### Problem: In a recent study, the Centers for Disease Control and Prevention reported that diastolic blood pressures of adult women in the United States are approximately normally distributed with mean 80.5 and standard deviation 9.9. ####

#### Graphing the normal distribution ####

```python
x = np.linspace(50, 120, 1000)
mean = 80.5
std_dev = 9.9
y = norm.pdf(x, loc=mean, scale=std_dev)
plt.fill_between(x, y, color='lightblue', alpha=0.5)
plt.plot(x,y)
```

#### What proportion of women have blood pressures lower than 70? ####

```python
plt.plot(x,y)
plt.fill_between(x, y, color='lightblue', alpha=0.5)

lower_than_70 = norm.cdf(70, loc=mean, scale=std_dev)

threshold = 70
x_shade = np.linspace(50, threshold, 1000)
y_shade = norm.pdf(x_shade, loc=mean, scale=std_dev)
plt.fill_between(x_shade, y_shade, color='purple', alpha=0.5)

print("The proportion of women that have blood pressue lower than 70 is {:.3f}".format(lower_than_70))
```

#### A diastolic blood pressure greater than 90 is classified as hypertension (high blood pressure). What proportion of women have hypertension? ####

```python
plt.plot(x,y)
plt.fill_between(x, y, color='lightblue', alpha=0.5)

higher_than_90 = 1 - norm.cdf(90, loc=mean, scale=std_dev)

threshold = 90
x_shade = np.linspace(threshold, 120, 1000)
y_shade = norm.pdf(x_shade, loc=mean, scale=std_dev)
plt.fill_between(x_shade, y_shade, color='purple', alpha=0.5)

print("The proportion of women that have blood pressue higher than 90 is {:.3f}".format(higher_than_90))
```
